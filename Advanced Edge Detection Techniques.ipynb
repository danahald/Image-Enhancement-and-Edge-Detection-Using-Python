```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('/content/image1.jpeg', cv2.IMREAD_GRAYSCALE)

# Sobel Edge Detection
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)


# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(sobelx**2 + sobely**2)

plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Sobel Edge Detection')
plt.show()


# Prewitt Operator
prewitt_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
prewitt_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])


# Apply Prewitt Operator
edge_x = cv2.filter2D(image, -1, prewitt_x)
edge_y = cv2.filter2D(image, -1, prewitt_y)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(edge_x, cmap='gray'), plt.title('Edge X')
plt.subplot(1, 2, 2), plt.imshow(edge_y, cmap='gray'), plt.title('Edge Y')
plt.show()


# Roberts operator kernels
roberts_cross_x = np.array([[1, 0],
                            [0, -1]])
roberts_cross_y = np.array([[0, 1],
                            [-1, 0]])

# Apply the Roberts operator kernels
edge_x = cv2.filter2D(image, -1, roberts_cross_x)
edge_y = cv2.filter2D(image, -1, roberts_cross_y)



plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(edge_x, cmap='gray'), plt.title('Roberts - Edge X')
plt.subplot(1, 2, 2), plt.imshow(edge_y, cmap='gray'), plt.title('Roberts - Edge Y')
plt.show()

```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python

image = cv2.imread('/content/image1.jpeg', cv2.IMREAD_GRAYSCALE)
# Apply Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (11, 11), 0)

# Sobel Edge Detection
sobelx = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=3)

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(sobelx**2 + sobely**2)

plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Sobel Edge Detection')
plt.show()


# Prewitt Operator
prewitt_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
prewitt_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

# Apply Prewitt Operator
edge_x = cv2.filter2D(blurred_image, -1, prewitt_x)
edge_y = cv2.filter2D(blurred_image, -1, prewitt_y)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(edge_x, cmap='gray'), plt.title('Edge X')
plt.subplot(1, 2, 2), plt.imshow(edge_y, cmap='gray'), plt.title('Edge Y')
plt.show()

# Roberts operator kernels
roberts_cross_x = np.array([[1, 0],
                            [0, -1]])
roberts_cross_y = np.array([[0, 1],
                            [-1, 0]])

# Apply the Roberts operator kernels
edge_x = cv2.filter2D(blurred_image, -1, roberts_cross_x)
edge_y = cv2.filter2D(blurred_image, -1, roberts_cross_y)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(edge_x, cmap='gray'), plt.title('Roberts - Edge X')
plt.subplot(1, 2, 2), plt.imshow(edge_y, cmap='gray'), plt.title('Roberts - Edge Y')
plt.show()
```
